```python
import pandas as pd
import sqlalchemy
```

```python
meteor_file = "resources/MeteoriteLanding.csv"
```

```python
meteor_file_df = pd.read_csv(meteor_file, encoding="ISO-8859-1")
```

```python
meteor_file_df.head()
```

```python
meteor_df = meteor_file_df[["ï»¿name", "id", "nametype", "recclass", "mass (g)", "fall", "year", "reclat", "reclong", "GeoLocation"]]
meteor_df.head()
```

```python
meteor_info = meteor_df.rename(columns={"ï»¿name":"name"})
meteor_info.head()
```

```python
df = meteor_info.set_index("id")
df.head()
```

```python
df = df.rename(columns={"mass (g)":"mass"})
df.head()
```

```python
print(df.dtypes)
```

```python
df['Mass'] = df.mass.astype('string')
df['Mass'] = df['mass'].replace(',','',regex=True, inplace=True)
df['Mass'] = df['mass'].astype(float)
df.head()
```

```python
df=df.loc[(df['Mass']>15)]
```

```python
df2 = df
```

```python
df2 = df2.dropna(how='any')
df2.head()
```

```python
df2.count()
```

```python
meteorite_df = df2[["name", "nametype", "recclass", "fall", "Mass", "year", "reclat", "reclong", "GeoLocation"]]
meteorite_df.head()
```

```python
meteorite_df.count()
```

```python
meteorite_df= meteorite_df.loc[(meteorite_df["reclat"]!=0)& (meteorite_df["reclong"]!=0)]
meteorite_df.head()
```

```python
meteorite_df.count()
```

```python
yucky_df = meteorite_df[meteorite_df['year'].str.contains("AM")]
yucky_df.head()
```

```python
yucky_df.count()
```

```python
meteorite_df = meteorite_df[~meteorite_df['year'].str.contains("AM")]
```

```python
final_meteorite_df = meteorite_df[["name", "recclass", "Mass", "year", "reclat", "reclong", "GeoLocation"]]
```
